```python
#### testing out a math problem animation
from manim import *
config.media_embed = True 
from manim import MathTex
param= "-v WARNING  --progress_bar None  -r  500,200  --disable_caching Example"
```

```python
%%manim $param

#%%manim -v WARNING  --progress_bar None -s -ql --disable_caching MyExample


class Example(Scene):
    def construct(self):
        mob=Text("Let's solve for x", color= YELLOW, slant=ITALIC)
        self.add(mob)
        self.wait(2)
        self.play(mob.animate.shift(2*UP))
        ul=Underline(mob, color=YELLOW)
        self.add(ul)
        self.play(ul.animate.scale(1.2))
        self.wait(1)
        

        mob2=MathTex("x^2 + 7x + 12=0", substrings_to_isolate="x", font_size=100)
        self.add(mob2)
        self.play(mob2.animate.set_color(BLUE))
        self.wait(1.5)
        self.play(FadeOut(ul,mob))
        self.play(FadeOut(mob2))
        

        mob3=Text("Step 1: Factor the equation")
        self.add(mob3)
        self.wait(1.5)
        self.play(mob3.animate.shift(2.2*UP).set_color(PINK).scale(0.8))
        self.wait(1.5)
        mob2_2=MathTex("x^2 + 7x + 12=0", substrings_to_isolate="x", font_size=70) 
        self.play(mob2_2.set_color(BLUE).animate.shift(1.35*UP))
        self.wait(1.5)

        arrow=Arrow(start=mob2_2.get_bottom(), end=DOWN, color=WHITE)
        self.play(GrowArrow(arrow))
        self.wait(1)

        mob4=MathTex("(x+3)(x+7)", substrings_to_isolate="x", font_size=60)
        mob4.move_to(arrow.get_bottom())
        self.play(mob4.animate.shift(0.7*DOWN))
        self.wait(2)

        self.play(FadeOut(mob2_2,mob3,arrow,mob4))

        mob5=Text("Step 2: Set the item in each set of parentheses equal to 0", font_size=36)
        self.add(mob5)
        self.wait(1.5)
        self.play(mob5.animate.shift(2.2*UP).set_color(PINK).scale(0.8))
        self.wait(1.5)

        mob4_2=MathTex("(x+3)(x+7)",font_size=60)
        self.add(mob4_2)
        self.wait(1)
        part_to_box1=VGroup(mob4_2[0][:5])
        box1=SurroundingRectangle(part_to_box1, color=YELLOW, buff=0.1)
        part_to_box2=VGroup(mob4_2[0][6:10])
        box2=SurroundingRectangle(part_to_box2, color=ORANGE, buff=0.1)
        self.add(box1)
        self.add(box2)
        self.wait(1)
        brack1=MathTex("x+3=0", font_size=60)
        brack2=MathTex("x+7=0",font_size=60)
        self.play(mob4_2.animate.shift(1.5*UP),box1.animate.shift(1.5*UP),box2.animate.shift(1.5*UP))
        self.wait(1)
        brack1.next_to(box1, DOWN, buff=1).shift(2*LEFT)
        brack2.next_to(box2, DOWN, buff=1).shift(2*RIGHT)
        arrow1=Arrow(start=box1.get_bottom(), end=brack1.get_top(), color=YELLOW)
        arrow2=Arrow(start=box2.get_bottom(), end=brack2.get_top(), color=ORANGE)
        self.play(FadeIn(brack1,brack2), GrowArrow(arrow1), GrowArrow(arrow2))
        self.wait(2)
        self.play(FadeOut(brack1,brack2,arrow1,arrow2,mob4_2,mob5,box1,box2))

        mob6=Text("Step 3: Solve for x")
        self.add(mob6)
        self.wait(1.5)
        self.play(mob6.animate.shift(2.2*UP).set_color(PINK).scale(0.8))
        self.wait(1.5)
        self.add((brack1).shift(0.5*LEFT),(brack2).shift(0.5*RIGHT))
        self.wait(1)
        self.play(brack1.animate.shift(1.5*UP).set_color(YELLOW),brack2.animate.shift(1.5*UP).set_color(ORANGE))
        self.wait(1)
        left1=MathTex("x+3-3=0-3", font_size=60)
        right1=MathTex("x+7-7=0-7",font_size=60)
        self.add(left1.next_to(brack1,DOWN,buff=0.5),right1.next_to(brack2,DOWN,buff=0.5))
        self.wait(2)
        left2=MathTex("x=-3", font_size=60)
        right2=MathTex("x=-7", font_size=60)
        
        
                 

        
        
```
